```python
# Extract the CSV from the zipfile
import zipfile
file = r"C:\Users\dyolf\netflix\netflix_titles.csv.zip"
with zipfile.ZipFile(file, mode='r', \
                     allowZip64=True, \
                     compresslevel=None, \
                     strict_timestamps=True,\
                     metadata_encoding=None) as zipref:
    with zipref.open('netflix_titles.csv', mode = 'r') as f:
        df = pd.read_csv(f, na_values = ['Not Available', 'unknown'])
```
